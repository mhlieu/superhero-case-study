# Statistical Analysis for Feature Selection

```python
from scipy.stats import chi2_contingency

categorical_columns = hero_df.select_dtypes(exclude = ['number', 'datetime']).columns

chi2_check = []
for i in categorical_columns:
    if chi2_contingency(pd.crosstab(hero_df['APPEARANCES'], hero_df[i]))[1] < 0.05:
        chi2_check.append('Reject Null Hypothesis')
    else:
        chi2_check.append('Fail to Reject Null Hypothesis')
res = pd.DataFrame(data = [categorical_columns, chi2_check] 
             ).T 
res.columns = ['Column', 'Hypothesis']
print(res)

categorical_columns
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import mutual_info_classif
from sklearn.feature_selection import chi2
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_med_appear, Y_med_appear, test_size=0.4, random_state=101)

fs = SelectKBest(score_func=chi2, k='all')
fs.fit(X_train, y_train)
X_train_fs = fs.transform(X_train)
X_test_fs = fs.transform(X_test)

category_names = X_med_appear.columns[fs.get_support(indices=True)].tolist()
fs_scores = []
for i in range(len(fs.scores_)):
    fs_scores.append(fs.scores_[i])
#     print((X.columns[fs.get_support(indices=True)][i], fs.scores_[i]))
#     print('Feature %d: %f' % (i, fs.scores_[i]))

fs_scores_df = pd.DataFrame(data = [category_names, fs_scores] 
             ).T 
fs_scores_df.columns = ['Category', 'Score']
fs_scores_df.sort_values(by='Category', ascending=True).to_clipboard()

# plot the scores
# plt.bar([i for i in range(len(fs.scores_))], fs.scores_)
# plt.show()
```
